#### Baseline Model Analysis
* The R squared score is 0.312, yet to not reliable
* t2, wind_speed, weather code and moths show high p-values (> 0.02)
* Skewness and Kurtosis are far from 0

```python
residuals = model.resid

fig, ax = plt.subplots(1,1)
fig.set_figheight(10)
fig.set_figwidth(10)

sm.ProbPlot(residuals).qqplot(line='s',ax=ax)
ax.title.set_text('Baseline/First model - QQ plot');
```

### 2-3. Third Modeling
#### 2-3-1. Log Transformation & Scaling - Normalising Data

```python
df3 = df2.copy()
```

```python
df3 = df3.reset_index()
```

```python
#conduct log transformation of the numeric values
a = df3['cnt']
b = df3['t1']
c = df3['hum']
d = df3['wind_speed']

logcnt = np.log(a, where = a>0)
logtemp = np.log(b, where = b>0)
loghum = np.log(c, where = c>0)
logwind = np.log(d, where = d>0)
```

```python
#conduct scaling of the numeric values
scaled_cnt = (logcnt-np.mean(logcnt))/np.sqrt(np.var(logcnt))
scaled_temp = (logtemp-np.mean(logtemp))/np.sqrt(np.var(logtemp))
scaled_hum = (loghum-np.mean(loghum))/np.sqrt(np.var(loghum))
scaled_wind = (logwind-np.mean(logwind))/np.sqrt(np.var(logwind))
```

```python
df3['cnt'] =scaled_cnt
df3['t1'] = scaled_temp
df3['hum'] = scaled_hum
df3['wind_speed'] = scaled_wind
```

```python
df3 = df3.drop('index', axis=1)
```

#### 2-3-2. Checking Modeling Result - Third Modeling

```python
outcome = 'cnt'
predictors = df3.drop('cnt', axis=1)
pred_sum = '+'.join(predictors.columns)
formula = outcome + '~' + pred_sum

model = ols(formula=formula, data=df3).fit()
model.summary()
```

#### Third Model Analysis
* The R-squared score has been improved to 0.807 from 0.611
* Skewness has been improved closer to 0 as -0.297
* Kurtosis also has been improved closer to 3
* Can see some of the variables are not statistically significant which is to be dropped for the final model

```python
residuals = model.resid

fig, ax = plt.subplots(1,1)
fig.set_figheight(10)
fig.set_figwidth(10)

sm.ProbPlot(residuals).qqplot(line='s',ax=ax)
ax.title.set_text('3rd model - QQ plot');
```

```python
y = df3[['cnt']]
X = df3.drop(['cnt'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)

linreg = LinearRegression()
linreg.fit(X_train, y_train)

y_hat_train = linreg.predict(X_train)
y_hat_test = linreg.predict(X_test)
train_prediction = linreg.predict(X_train) 
test_prediction = linreg.predict(X_test)

plt.figure(figsize=(8,5))
plt.scatter(y_train, train_prediction, label='Model')
plt.plot(y_train, y_train, label='Actual Data')
plt.title('Model vs Data for training set')
plt.legend();
#Can observe some outliers
```
